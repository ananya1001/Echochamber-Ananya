```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
years=[2014,2015,2016,2017,2018,2019,2020,2021]
income=[55,56,62,61,72,72,73,75]
income_ticks=list(range(50,81,2))
plt.plot(years,income)
plt.title("Income of John in USD",fontsize=10, fontname="Monstserrat")
plt.xlabel("year")
plt.ylabel("Income in USD")
plt.yticks(income_ticks, [f"{x}k USD" for x in income_ticks])
plt.show()
```

```python
stock_a=[100,90,102,99,101,100]
stock_=[101,99,100,90,102,110]
stock_c=[111,115,100,90,95,100]
plt.plot(stock_a,label="stock_a")
plt.plot(stock_,label="stock_b")
plt.plot(stock_c,label="stock_c")
plt.legend(loc="upper center")
plt.show()
```

```python
votes=[10,2,5,16,22]
people=["A","B","C","D","E"]
plt.pie(votes,labels=None)
plt.legend(labels=people,loc="lower left",fontsize=7)
plt.show()
```

```python
from matplotlib import style
style.use("dark_background")
votes=[10,2,5,16,22]
people=["A","B","C","D","E"]
plt.pie(votes,labels=None)
plt.legend(labels=people,loc="lower left",fontsize=7)
plt.show()
```

```python
x1,y1=np.random.random(100),np.random.random(100)
x2,y2=np.arange(100), np.random.random(100)
plt.figure(1)
plt.scatter(x1,y1)
plt.figure(2)
plt.plot(x2,y2)
plt.show()
```

```python
x=np.arange(100)
fig, axs=plt.subplots(2,2)
axs[0,0].plot(x,np.sin(x))
axs[0,0].set_title("sin(x)")
axs[0,1].plot(x,np.cos(x))
axs[1,0].plot(x,np.tan(x))
axs[1,1].plot(x,np.log(x))
plt.savefig("test.png",dpi=300,transparent=True)
plt.tight_layout()
plt.show()
```

```python
ax=plt.axes(projection="3d")

x=np.arange(0,50,0.1)
y=np.sin(x)
z=np.cos(x)
ax.plot(x,y,z)
```
